#### Q 5. Find the net revenue generated by the E-Commerce company over all orders placed.

```python
E_com_rev=0
for k in ECom_Data['discounted_price']:
    if (k>600):
        E_com_rev = E_com_rev + 0.25*k
    elif (k>350) & (k<=600):
        E_com_rev = E_com_rev + 0.15*k
    elif (k>100) & (k<=350):
        E_com_rev = E_com_rev + 0.10*k
    else:
        E_com_rev = E_com_rev + 0.05*k
print(f'net revenue generated by the E-Commerce company over all orders placed:', {E_com_rev})
```

#### Q6.Calculate the BrandRevenue for each brand and list the top 14 brands having maximum revenue in descending order (6 marks)
* Hint: Brand revenue is different from the E-Commerce company revenue.

```python
# It will give the company's charge on each brand
lst=[]
for k in ECom_Data['discounted_price']:
    if (k>600):
        E_com_rev1 =  0.25*k
    elif (k>350) & (k<=600):
        E_com_rev1 = 0.15*k
    elif (k>100) & (k<=350):
        E_com_rev1 = 0.10*k
    else:
        E_com_rev1 = 0.05*k
    lst.append(E_com_rev1)
lst
```

```python
# It will create one another column of E_com_rev1
ECom_Data['E_com_rev1'] = lst
ECom_Data.head(4)
```

```python
# It will give the different column for BrandRevenue which is (Total selling price - flip kart charged)
ECom_Data['BrandRevenue'] = ECom_Data['discounted_price'] - ECom_Data['E_com_rev1']
ECom_Data.head(14)
```

```python
# It will give total brand Revenue for each brand
brand_Revenue = ECom_Data.groupby('Brand')[['BrandRevenue']].sum().sort_values('BrandRevenue',ascending = False).head(14)
brand_Revenue
```

#### Q 8. Create a new dataframe to include the brand specific information as stated:
* i. total number of orders placed per brand
* ii. total retail_price per brand
* iii. total discount_price per brand, and
* iv. total BrandRevenue generated per brand.
* Also, draw a pairplot using these four features. (6 marks)

```python
# total number of orders placed per brand
Total_no_order_per_brand = (ECom_Data.value_counts(['Brand']))
Total_no_order_per_brand = Total_no_order_per_brand.to_frame().reset_index()
Total_no_order_per_brand.columns = ['Brand', 'order']
Total_no_order_per_brand
```

```python
# total retail_price per brand
Total_ret_per_Brand = (ECom_Data.groupby(['Brand'])['retail_price'].sum())
Total_ret_per_Brand = Total_ret_per_Brand.to_frame().reset_index()
Total_ret_per_Brand.columns = ['Brand', 'retail_price']
Total_ret_per_Brand
```

```python
a = Total_no_order_per_brand.merge(Total_ret_per_Brand)
a
```

```python
#  total discount_price per brand
Total_dis_per_Brand = (ECom_Data.groupby(['Brand'])['discounted_price'].sum())
Total_dis_per_Brand = Total_dis_per_Brand.to_frame().reset_index()
Total_dis_per_Brand.columns = ['Brand', 'discounted_price']
Total_dis_per_Brand
```

```python
b = a.merge(Total_dis_per_Brand)
b
```

```python
# total BrandRevenue generated per brand.
Total_brand_rev_per_Brand = (ECom_Data.groupby(['Brand'])['BrandRevenue'].sum())
Total_brand_rev_per_Brand = Total_brand_rev_per_Brand.to_frame().reset_index()
Total_brand_rev_per_Brand.columns = ['Brand', 'BrandRevenue']
Total_brand_rev_per_Brand
```

```python
c = b.merge(Total_brand_rev_per_Brand)
c
```

```python
sns.pairplot(data=c);
```

#### Q 9.Compare performance regionwise
* a) Draw a lineplot for the monthly Revenue of ECom Company for each region separately. (4 marks)
* b) Identify the best and the worst performing months for each region. (2 marks)
* Note: Only those days with actual orders placed are present in the dataset. Assuming there were no orders on other days


```python
ECom_Data['Order_Date'] = pd.to_datetime(ECom_Data['Order_Date'],dayfirst=True)
```

```python
ECom_Data['Year'] = ECom_Data['Order_Date'].dt.year
ECom_Data['Month'] = ECom_Data['Order_Date'].dt.month
ECom_Data['Day'] = ECom_Data['Order_Date'].dt.day
ECom_Data['year_month'] = ECom_Data['Order_Date'].dt.strftime("%Y-%m")
```

```python
group_all = ECom_Data.groupby(['Region','Year','Month','year_month'], as_index=False).sum()
group_all
```

```python
plt.figure(figsize=(40,20))
sns.lineplot(data=group_all,x='year_month',y='E_com_rev1',hue='Region');
```

```python
worst_reg = group_all.groupby(['Region'])[['E_com_rev1']].min()
worst_reg
```

```python
pd.merge(group_all,worst_reg,on = ['Region','E_com_rev1'])

# October 2021 is the worst performing month for East region.
# July 2021 is the worst performing month for North region.
# May 2021 is the worst performing month for South region.
# October 2021 is the worst performing month for West region.
```

```python
best_reg = group_all.groupby(['Region'])[['E_com_rev1']].max()
best_reg
```

```python
pd.merge(group_all,best_reg,on = ['Region','E_com_rev1'], how = 'inner')

# December 2020 was the best performing month for all region.
```
